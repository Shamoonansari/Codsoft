```python
import pandas as pd
import numpy as np
from pandas.plotting import scatter_matrix

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("IRIS.csv")
```

```python
df
```

```python
df.info()
```

```python
df["species"].value_counts()
```

Now we know that there are 3 species

```python
df["species_id"]=df["species"]
df["species_id"].replace({"Iris-setosa":0,"Iris-versicolor":1,"Iris-virginica":2},inplace=True)
```

```python
df
```

# EDA

```python
fig=plt.figure()
scatter_matrix(df,figsize=(10,10),alpha=0.9,diagonal="kde",marker="o")
plt.show()
```

# Data Splitting

```python
x=df["species_id"]
y=df.drop(["species","species_id"],axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(y, x, test_size=0.2)
```

# Model Selection and Training

we will use DecisionTreeClasification

```python
clf = DecisionTreeClassifier()

clf.fit(x_train, y_train)
```

```python


# Make predictions on the test data
y_pred = clf.predict(x_test)

# Evaluate the model's accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```
